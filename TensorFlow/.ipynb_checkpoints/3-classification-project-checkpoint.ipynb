```python
y = data_frame["Type"].values
x = data_frame.drop("Type", axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.3, random_state= 15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()
model.add(Dense(units = 30, activation ="relu"))
model.add(Dropout(0.6))
model.add(Dense(units = 15, activation ="relu"))
model.add(Dropout(0.6))
model.add(Dense(units = 15, activation ="relu"))
model.add(Dropout(0.6))
model.add(Dense(units = 1, activation ="sigmoid"))

model.compile(loss = "binary_crossentropy", optimizer= "adam")
```

```python
early_stopping = EarlyStopping(monitor="val_loss", mode="min", verbose=1, patience=35)
```

```python
model.fit(x_train, y_train, epochs=700, validation_data = (x_test, y_test), verbose = 1, callbacks= [early_stopping])
```

```python
modelKaybı = pd.DataFrame(model.history.history)
```

```python
modelKaybı.plot()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
threshold = 0.5
y_pred_classes = (model.predict(x_test) > threshold).astype(int)
```

```python
print(classification_report(y_test, y_pred_classes))
```

```python
confusion_matrix(y_test, y_pred_classes)
```
