```python
data_frame = yuzde_doksan_dokuz_df
```

```python
data_frame =  data_frame[data_frame["year"] != 1970]
```

```python
data_frame.head()
```

```python
data_frame = data_frame.drop("transmission", axis = 1)
```

```python
y = data_frame["price"].values
x = data_frame.drop("price", axis= 1).values
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.3, random_state= 10) 
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(12, activation= "relu"))
model.add(Dense(12, activation= "relu"))
model.add(Dense(12, activation= "relu"))
model.add(Dense(12, activation= "relu"))

model.add(Dense(1))

model.compile(optimizer="adam", loss = "mse")
```

```python
model.fit(x_train, y_train, validation_data=(x_test, y_test), batch_size= 250, epochs = 300)
```

```python
kayÄ±p_verisi = pd.DataFrame(model.history.history)
```

```python
kayÄ±p_verisi.head()
```

```python
kayÄ±p_verisi.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
tahmin_dizisi = model.predict(x_test)
```

```python
tahmin_dizisi
```

```python
mean_absolute_error(y_test, tahmin_dizisi)
```
