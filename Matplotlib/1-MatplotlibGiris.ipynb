# Plot() Method'u

Matplotlib kütüphanesinde plot() metodu, 2D grafikler çizmek için kullanılır. En sık kullanılan grafik türü çizgi grafiktir, ancak bu metodla diğer grafik türlerini de oluşturabilirsiniz.

## Subplot()

`subplot()` metodu, Matplotlib kütüphanesinde aynı figür içinde birden fazla grafik oluşturmanızı sağlar. Grafiklerin düzenini satır ve sütunlar halinde belirlemenize olanak tanır. İşte `subplot()` metodunun nasıl kullanıldığına dair detaylar:

### `subplot()` Metodunun Parametreleri

- **`nrows`**: Grafiklerin yerleştirileceği satır sayısını belirtir.
- **`ncols`**: Grafiklerin yerleştirileceği sütun sayısını belirtir.
- **`index`**: Grafiğin figürdeki konumunu belirtir. Bu değer, 1'den başlayarak artar ve soldan sağa, yukarıdan aşağıya doğru sıralanır.

````python
plt.subplot(1,2,1) # 1 sıra olacak, 2 kolon olacak, suan 1. grafiği
plt.plot(numpy_dizisi1, numpy_dizisi2, "r*-")
plt.subplot(1,2,2)  # 1 sıra olacak, 2 kolon olacak, suan 2. grafiği
plt.plot(numpy_dizisi2, numpy_dizisi1, "g--")
plt.show()
````

## Extra Örnek:

````python
import matplotlib.pyplot as plt

# İlk alt grafik (2 satır, 2 sütun, 1. konum)
plt.subplot(2, 2, 1)  # 2 satır, 2 sütun, 1. konum
plt.plot([1, 2, 3], [1, 4, 9])
plt.title('Grafik 1')

# İkinci alt grafik (2 satır, 2 sütun, 2. konum)
plt.subplot(2, 2, 2)  # 2 satır, 2 sütun, 2. konum
plt.plot([1, 2, 3], [9, 4, 1])
plt.title('Grafik 2')

# Üçüncü alt grafik (2 satır, 2 sütun, 3. konum)
plt.subplot(2, 2, 3)  # 2 satır, 2 sütun, 3. konum
plt.plot([1, 2, 3], [2, 5, 3])
plt.title('Grafik 3')

# Dördüncü alt grafik (2 satır, 2 sütun, 4. konum)
plt.subplot(2, 2, 4)  # 2 satır, 2 sütun, 4. konum
plt.plot([1, 2, 3], [4, 6, 2])
plt.title('Grafik 4')

# Alt grafikleri düzenleme
plt.tight_layout()

plt.show()
````

## Figure()

`matplotlib.figure()` metodu, `matplotlib` kütüphanesinde grafik çizmek için kullanılan bir figür (tuval) oluşturur. Bu figür, bir veya birden fazla grafik ekseni (`axes`) içerebilir ve grafiksel elemanların (çizgiler, noktalar, metinler vb.) üzerine çizileceği ana kapsayıcı alanı temsil eder.

````python
## Tek tek açıklayalımm
````

````python
benim_figure = plt.figure()
figure_axes = benim_figure.add_axes([0.2, 0.2, 0.9, 0.9])
````

**Eksenin Genişliği:**

Figürün genişliği 10 inç. Eksenin genişliği, figür genişliğinin %90'ı (0.9) olduğuna göre: <br>
10 inç × 0.9 = 9 inç 

Yani eksen, figürün genişliğinin 9 inçlik kısmını kaplayacaktır.

**Eksenin Yüksekliği:**

Figürün yüksekliği 10 inç. Eksenin yüksekliği, figür yüksekliğinin %90'ı (0.9) olduğuna  <br>
10 inç × 0.9 = 9 inç <br><br>g
Not: add_axes([0.2, 0.2, weight, heigt] bak bizim grafiğimiz toplamda 10 inç o yüzden 0 ila 1 arasında değerler vererek 
grafiğimizi oluşturabiliriz figure() method'uda bize çalışabileceğimiz bir alan verirre:



````python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1, 0.1, 0.7, 0.7])
eksen2 = figure2.add_axes([0.3, 0.3, 0.3, 0.3])

eksen1.plot(numpy_dizisi1, numpy_dizisi2, "g")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpy_dizisi2, numpy_dizisi1, "b")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("Y ekseni")
eksen2.set_title("Küçük Grafik Başlık")
````
